# **Auto Completion Tool**

```python
def print_distances(distances,word1_length,word2_length):
  """pair each number starting from zero ending with the length of word1 +1
  with each number starting from zero and ending with the length of word2 +1
  and use those pairs as indices to print out values from the distances matrix"""
  for w1 in range(word1_length+1):
    for w2 in range(word2_length+1):
      print(int(distances[w1][w2]),end = ' ')
    print()
```

```python
def levenshteinDistanceDP(token1, token2):
  # create a matrix of size m,n with m being the length of word 1
  # and n being the length of word 2.  The matrix will start as all zeroes
    distances = np.zeros((len(token1) + 1, len(token2) + 1))

  #the rows will be labeled 0 through len(token1) and the same for the the columns
  # but with word 2
    for t1 in range(len(token1) + 1):
        distances[t1][0] = t1

    for t2 in range(len(token2) + 1):
        distances[0][t2] = t2

    a = 0
    b = 0
    c = 0

    # for x(t1,t2) in the matrix, compare the t1th letter of word 1 to the
    # t2th letter of word 2. If they are equal, the value of x(t1,t2) is equal
    # to the value of x(t1-1,xt2-1), which is up and to the left.
    for t1 in range(1, len(token1) + 1):
        for t2 in range(1, len(token2) + 1):
            if (token1[t1-1] == token2[t2-1]):
                distances[t1][t2] = distances[t1 - 1][t2 - 1]
            else:
                a = distances[t1][t2 - 1]
                b = distances[t1 - 1][t2]
                c = distances[t1 - 1][t2 - 1]

                # otherwise, select the minimum value from below and add 1
                if (a <= b and a <= c):
                    distances[t1][t2] = a + 1
                elif (b <= a and b <= c):
                    distances[t1][t2] = b + 1
                else:
                    distances[t1][t2] = c + 1

    return distances[len(token1)][len(token2)]
```

```python
def calcDictDistance(word, numWords):
  # opens file
    file = open('WordList.txt', 'r', errors = 'ignore')
    # accesses lines in file
    lines = file.readlines()
    file.close()
    dictWordDist = []
    wordIdx = 0


    for line in lines:
        if ' ' in word:
          word = word.split(' ')
          word = word[-1]
          # line.strip() is creating word2. Because the line is a word from word
          # list, all that needs to happen is remove the new line and any space
        wordDistance = levenshteinDistanceDP(word, line.strip())
        if wordDistance >= 10:
            wordDistance = 9
            # format is distance - word2
        dictWordDist.append(str(int(wordDistance)) + "-" + line.strip())
        wordIdx = wordIdx + 1

    closestWords = []
    wordDetails = []
    currWordDist = 0
    dictWordDist.sort()
    print(dictWordDist)
    for i in range(numWords):
        currWordDist = dictWordDist[i]
        wordDetails = currWordDist.split("-")
        closestWords.append(wordDetails[1])
    return closestWords
```
